```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def diff_equation1(V1,t1):
    return -10 - V1 - t1

V1 = np.arange(-20, 6, 2)
t1 = np.arange(-6, 10, 0.5)
T1, V1 = np.meshgrid(V1, t1)

slope_field_diff_equation1 = diff_equation1(V1, T1)

plt.quiver(V1, T1, np.ones_like(slope_field_diff_equation1), slope_field_diff_equation1, color='b', scale=8, scale_units='xy', angles='xy', label='Slope Field')
contour = plt.contour(V1, T1, slope_field_diff_equation1, levels=[-3, -1, 1], colors=['b', 'g', 'm'], linewidths=1)
plt.title('Differential Equation 1 slope field')
plt.xlabel('t1(s)')
plt.ylabel('V1(V)')
plt.legend(loc="upper right")
plt.savefig("Differential Equation 1.pdf")
plt.show()
```

```python
def diff_equation2(V2,t2):
    return np.cos(t2) - 0.5 * V2 + 20
V2 = np.arange(30, 60, 2)
t2 = np.arange(0, 10, 0.5)
T2, V2 = np.meshgrid(V2, t2)

slope_field_diff_equation2 = diff_equation2(T2, V2)

plt.quiver(V2, T2, np.ones_like(slope_field_diff_equation2), slope_field_diff_equation2, color='b', scale=5, scale_units='xy', angles='xy', label='Slope Field')
contour = plt.contour(V2, T2, slope_field_diff_equation2, levels=[-2, 0, 2], colors=['g', 'm', 'c'], linewidths=1)
plt.title('Slope Field for Equation 2 with Isoclines')
plt.xlabel('t2(s)')
plt.ylabel('V2(V)')
plt.legend(loc="upper right")
plt.savefig("Differential Equation 2.pdf")
plt.show()
```

```python
def diff_equation3(t3, V3):
    return Imax1 * np.sin(t3) / Cm1 - (V3 / (R1 * Cm1))

R1 = 1  
Cm1 = 2  
Imax1 = 0  

V3 = np.arange(-20, 25, 2)
t3 = np.arange(0, 6, 0.5)
T3, V3 = np.meshgrid(t3, V3)

slope_field_eq3 = diff_equation3(t3, V3)

plt.figure(figsize=(8, 6))
plt.quiver(T3, V3, np.ones_like(slope_field_eq3), slope_field_eq3, color='b', scale=5, scale_units='xy', angles='xy',label='Slope Field Imax = 0')
plt.title('Slope Field for Cell Circuit')
plt.xlabel('t(s)')
plt.ylabel('V(V)')
plt.legend(loc="upper right")
plt.savefig("Slope Field Cell Circuit 1.pdf")
plt.show()
```

```python
def diff_equation4(t3, V4):
    return Imax2 * np.sin(t3) / Cm2 - (V3 / (R2 * Cm2))

R2 = 1  
Cm2= 2  
Imax2 = 10  

slope_field_eq4 = diff_equation4(t3, V3)

plt.quiver(T3, V3, np.ones_like(slope_field_eq4), slope_field_eq4, color='b', scale=5, scale_units='xy', angles='xy',label='Slope Field Imax=10')
plt.title('Slope Field for Cell Circuit')
plt.xlabel('t(s)')
plt.ylabel('V(V)')
plt.legend(loc="upper right")
plt.savefig("Slope Field Cell Circuit 2.pdf")
plt.show()
```
